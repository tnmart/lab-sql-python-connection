2. Write a Python function called rentals_month that retrieves rental data for a given month and year (passed as parameters) from the Sakila database as a Pandas DataFrame. The function should take in three parameters:

engine: an object representing the database connection engine to be used to establish a connection to the Sakila database.

month: an integer representing the month for which rental data is to be retrieved.

year: an integer representing the year for which rental data is to be retrieved.

The function should execute a SQL query to retrieve the rental data for the specified month and year from the rental table in the Sakila database, and return it as a pandas DataFrame.

```python
def rentals_month(engine_name:str, month_name: str, year: int)-> pd.DataFrame:
    '''
    This function retrieves data from the Sakila database regarding the rentals in a given month and year and outputs
    it into a pandas dataframe.
    Inputs:
    engine: an object representing the database connection engine to be used to establish a connection to the Sakila database
    month: an integer representing the month for which rental data is to be retrieved.
    year: an integer representing the year for which rental data is to be retrieved.
    Output: rental data for a given month and year in a Pandas DataFrame
    '''
    query = f"""SELECT *,
    YEAR(rental_date) AS year_of_rental,
    MONTHNAME(rental_date) AS month_of_rental
    FROM sakila.rental
    WHERE YEAR(rental_date) = {year} AND MONTHNAME(rental_date) = '{month_name}';
    """
    result = pd.read_sql_query(query, engine_name)
    return result

rentals_month(engine, "June", 2005)
```

3. Develop a Python function called rental_count_month that takes the DataFrame provided by rentals_month as input along with the month and year and returns a new DataFrame containing the number of rentals made by each customer_id during the selected month and year.

The function should also include the month and year as parameters and use them to name the new column according to the month and year, for example, if the input month is 05 and the year is 2005, the column name should be "rentals_05_2005".

Hint: Consider making use of pandas groupby()

```python
def rental_count_month (rental_month_df: pd.DataFrame, colname_month: str, colname_year: str)->pd.DataFrame:
    '''
    Takes the rentals_month table as a pandas dataframe along with the month and year and returns a new DataFrame
    containing the number of rentals made by each customer_id during the selected month and year
    Inputs:
    rental_month_df - pd.DataFrame
    colname_month - str
    colname_year - str
    Output:
    rentals_groupby - pd.DataFrame
    '''
    rentals_groupby = rental_month_df.groupby(['customer_id','year_of_rental','month_of_rental']).agg({"rental_date": 'count'}).reset_index()
    rentals_groupby.columns.values[3] = f'rentals_{colname_month}_{colname_year}'
    return rentals_groupby


rental_count_month(rentals_month(engine, "May", 2005), "May", 2005)
```

4. Create a Python function called compare_rentals that takes two DataFrames as input containing the number of rentals made by each customer in different months and years. The function should return a combined DataFrame with a new 'difference' column, which is the difference between the number of rentals in the two months.

```python
def compare_rentals(dataframe1: pd.DataFrame, dataframe2: pd.DataFrame, month1: str, year1: str, month2: str, year2: str)->pd.DataFrame:
    '''
    Takes 2 pandas dataframes with the number of rentals made by each customer in different months and years
    and returns a dataframe with an added column showing the difference in rentals between the two columns.
    Inputs:
    dataframe1 - pd.DataFrame
    dataframe2 - pd.DataFrame
    month1 - str
    year1 - str
    month2 - str
    year2 - str
    Output:
    dataframe3 - pd.DataFrame
    '''
    dataframe3=dataframe1.merge(dataframe2, how="inner", on="customer_id")
    dataframe3["difference"] = dataframe3[f'rentals_{month1}_{year1}']-dataframe3[f'rentals_{month2}_{year2}']
    return dataframe3

df1 = rental_count_month(rentals_month(engine, "May", 2005), "May", 2005)
df2 = rental_count_month(rentals_month(engine, "June", 2005), "June", 2005)
compare_rentals(df1, df2, "May", 2005, "June", 2005)
```
